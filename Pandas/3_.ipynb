## <font color='blue'>Using Merge Operations</font>

```python
data4 = {'emp_id' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
```

```python
data5  = {'emp_id' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
    
}
```

```python
df4 = pd.DataFrame(data4)
```

```python
df5 = pd.DataFrame(data5)
```

```python
df4
```

```python
df5
```

```python
pd.merge(df4,df5)    # it will merge common things
```

### <font color='blue'>- join operations</font>

```python
# left join

pd.merge(df4,df5, how='left')
```

```python
# right join

pd.merge(df4,df5, how='right')
```

### <font color='blue'>- using ON operations</font>

```python
# using on

pd.merge(df4,df5, how='right', on='emp_id')
```

```python
data6 = {'emp_id1' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
         
data7  = {'emp_id2' : [101,102,104,105],
          'mob_no' : [3434,5656,767,878],
          'house_nu' : [345,345,345,56]
    
}
```

```python
df6 = pd.DataFrame(data6)
df7= pd.DataFrame(data7)
```

```python
df6
```

```python
df7
```

```python
pd.merge(df6,df7)   # there is no common column thats y this will showing u error
```

```python
# if no column name is match on both table then we have to use on operations like left_on, right_on

pd.merge(df6,df7, left_on='emp_id1',right_on='emp_id2')
```

## <font color='blue'>Function apply in the dataset</font>

```python
df_sales = pd.read_csv("C:/Users/raman/Downloads/sales_data_final.csv")
```

```python
df_sales.head()
```

### - function created

```python
def profit_flag(a):
    if a>=0:
        return("Positive")
    else:
        return("Negative")
```

```python
profit_flag(45)
```

```python
profit_flag(-45)
```

### - How to apply the functions in the dataset just using apply

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag) # using apply
```

```python
df_sales.head()
```

```python
# apply len functions on the customer name

df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
# create a flag for the quantity
# if quantity b/w 0-10 it will low
# if quantity b/w 11-20 it will be medium
# if quantity above 20 it will be high


def quantity_flag(a):
    if a<=10:
        return("Low")
    elif a<=20:
        return("Medium")
    else:
        return("High")
```

```python
df_sales['flag_quantity'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
df_sales[df_sales.flag_quantity == 'Low'].count()
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a:a**2)
```
